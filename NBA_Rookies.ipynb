 ## Importing the essential libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

## Loading the training data

```python
training_data = pd.read_csv('Train_data.csv')
X1 = training_data.iloc[:, :-1].values
y1 = training_data.iloc[:, -1].values
```

```python
training_data.head(10)
```

```python
testing_data = pd.read_csv('Test_data.csv')
```

```python
testing_data.head(10)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X1 = sc.fit_transform(X1)
X2 = sc.transform(testing_data) 
```

```python
print(X1)
```

```python
print(X2)
```

## Modeling using Support Vector Classifier

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', gamma=0.5, C = 1000)
classifier.fit(X1, y1)
```

## Making Predictions and saving it in a variable

```python
y_pred = classifier.predict(X2)
```

```python
print(y_pred)
```

```python
# Total number of values predicted
np.count_nonzero(y_pred == 0) + np.count_nonzero(y_pred) 
```

```python
prediction = pd.DataFrame(y_pred, columns=['prediction']).to_csv('prediction.csv')
```
